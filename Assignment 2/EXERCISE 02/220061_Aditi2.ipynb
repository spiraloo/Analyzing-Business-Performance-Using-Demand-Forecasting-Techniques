# Exercises

**Recreate the plots below using the titanic dataframe. There are very few hints since most of the plots can be done with just one or two lines of code and a hint would basically give away the solution. Keep careful attention to the x and y labels for hints.**

*Note! In order to not lose the plot image, make sure you don't code in the cell that is directly above the plot, there is an extra cell above that one which won't overwrite that plot!*

```python
import numpy as np
import matplotlib.pyplot as plt
fig=sns.histplot(data=titanic,x='fare',bins=np.arange(0,600,(100/6)))
fig.set(ylabel='')
plt.xlim(0,600)
```

```python
fig=sns.boxplot(x='class',y='age',data=titanic)
fig.set(ylim=(0,80))
```

```python
import matplotlib.pyplot as plt
sns.swarmplot(data=titanic,x='class',y='age',hue='class',legend=False)
plt.yticks([0,20,40,60,80])
```

```python
sns.countplot(x="sex",data=titanic)
```

```python
corr=titanic.corr()
fig=sns.heatmap(corr,vmin=-0.8,vmax=0.8)
fig.set_xticklabels(fig.get_xticklabels(), rotation=0)
fig
plt.title('titanic.corr()')
```

```python
import matplotlib.pyplot as plt
fig=sns.FacetGrid(data=titanic,col="sex")
fig.map(sns.histplot,"age",bins=10)
fig.set(ylabel=(''),xlim=(0,80))
plt.xticks([0,10,20,30,40,50,60,70,80])
```
